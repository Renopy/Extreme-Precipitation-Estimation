```python
fp <- 'H:/Projects/USA/data/resample/USA_2019-2023_resmaple_12.csv' 

df = read.csv(fp)

col_name = names(df)
```

```python
df = filter(df, V != -8888 )  
df = filter(df, PWV_ave  < 1000 )  
df = filter(df, t2m   > -30000 )  
```

```python
ggplot(data = df, mapping = aes(x = PWV_ave, y = V )) +
  geom_boxplot(fill = "bisque",color = "black", alpha = 0.3) +
  geom_jitter(aes(color = 'blue'), alpha=0.2) +
  labs(x = "PWV") +
  ggtitle("Vapor index") +
  guides(color = FALSE) +
  theme_minimal()  
```

```python

ggplot(data = df, mapping = aes(x = PWV_ave, y = t2m)) +
   geom_point() +
   ggtitle("Alaska Flight Depature Delays vs Arrival Delays")
```

```python
ggplot(data = df, mapping = aes(x = PWV_ave, y = t2m)) +
    geom_point() + 
    geom_smooth(method = "lm", na.rm = TRUE)
print( cor( df$PWV_ave , df$t2m  ))
```
